```python
df = []
for i in range(3):
    df.append(pd.read_csv(os.path.join("data/", cab_data[i])))
```

```python
res = []
for i in range(3):
    res_list = []
    for index, row in df[i].iterrows():
        if(index % 10000 == 0):
            print(index)
    
        pickup_datetime = datetime.datetime.strptime(row['tpep_pickup_datetime'], '%Y-%m-%d %H:%M:%S')
        dropoff_datetime = datetime.datetime.strptime(row['tpep_dropoff_datetime'], '%Y-%m-%d %H:%M:%S')

        res_list.append({'pickup_date' : row['tpep_pickup_datetime'][:-3], 
                    'pickup_weekday': pickup_datetime.weekday(),
                    'pickup_hour': pickup_datetime.hour,
                    'pickup_min': pickup_datetime.minute,
                    'trip_distance': float(row['trip_distance']),
                    'PULocationID': int(row['PULocationID']),
                    'fare_amt': float(row['fare_amount']),
                    'total_profit': float(row['total_amount'])
                   })
    res.append(res_list)
```

```python
df_output = []
for i in range(3):
    df_output.append(pd.DataFrame(res[i]))
```

```python
df_reset = []
for i in range(3):
    df_reset.append(df_output[i].groupby(['pickup_date', 'pickup_hour', 'pickup_min', 'PULocationID', 'pickup_weekday'])['total_profit'].mean().reset_index())
```

```python

df_reset0 = df_reset[0][("2018-01-01 00:00" <= df_reset[0]['pickup_date']) & (df_reset[0]['pickup_date'] < "2018-02-01 00:00")].reset_index()
df_reset1 = df_reset[1][("2018-02-01 00:00" <= df_reset[1]['pickup_date']) & (df_reset[1]['pickup_date'] < "2018-03-01 00:00")].reset_index()
df_reset2 = df_reset[2][("2018-03-01 00:00" <= df_reset[2]['pickup_date']) & (df_reset[2]['pickup_date'] < "2018-04-01 00:00")].reset_index()
```

```python
print(df_reset0)
print(df_reset1)
print(df_reset2)
# print(df_reset[2])
```

```python
df_final = df_reset0.append(df_reset1, ignore_index = True)
df_final = df_final.append(df_reset2, ignore_index = True)
df_final
```

```python
df_final.to_pickle('processed_data.pkl')
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
import numpy
```

```python
X = df_output.as_matrix(columns=['pickup_weekday', 'pickup_hour', 'pickup_min','PULocationID'])
y = df_output.as_matrix(columns=['total_profit']).ravel()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
model = RandomForestRegressor(n_estimators=20, verbose=2)
model.fit(X_train, y_train)
```

```python
r_square = model.score(X_train, y_train)
```

```python
r_test = model.score(X_test, y_test)
```
