```python
df = pd.read_csv('data/yellow_tripdata_2018-01.csv')
```

```python
print(len(df.index))
```

```python
datetime.datetime.strptime('2018-01-02 12:30:00', '%Y-%m-%d %H:%M:%S').minute
```

```python
res = []
for index, row in df.iterrows():
    if(index % 10000 == 0):
        print(index)
    
    pickup_datetime = datetime.datetime.strptime(row['tpep_pickup_datetime'], '%Y-%m-%d %H:%M:%S')
    dropoff_datetime = datetime.datetime.strptime(row['tpep_dropoff_datetime'], '%Y-%m-%d %H:%M:%S')
    
    res.append({'pickup_weekday': pickup_datetime.weekday(),
                'dropoff_weekday': dropoff_datetime.weekday(),
                'pickup_hour': pickup_datetime.hour,
                'dropoff_hour': dropoff_datetime.hour,
                'pickup_min': pickup_datetime.minute,
                'dropoff_min': dropoff_datetime.minute,
                'trip_distance': float(row['trip_distance']),
                'PULocationID': int(row['PULocationID']),
                'DOLocationID': int(row['DOLocationID']),
                'fare_amt': float(row['fare_amount']),
                'total_profit': float(row['total_amount'])
               })
```

```python
df_output = pd.DataFrame(res)
```

```python
print(len(df_output.index))
```

```python
df_output['PULocationID']
```

```python
df_output.to_pickle('processed_data.pkl')
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
import numpy
```

```python
X = df_output.as_matrix(columns=['pickup_weekday', 'pickup_hour', 'pickup_min','PULocationID'])
y = df_output.as_matrix(columns=['total_profit']).ravel()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
model = RandomForestRegressor(n_estimators=20, verbose=2)
model.fit(X_train, y_train)
```

```python
r_square = model.score(X_train, y_train)
```

```python
r_square
```

```python
r_test = model.score(X_test, y_test)
```

```python
r_test
```
