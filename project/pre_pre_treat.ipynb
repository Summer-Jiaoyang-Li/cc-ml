```python
import pandas as pd
import datetime
```

```python
import csv
import numpy as np

res = []

with open('data/yellow_tripdata_2018-03.csv', mode='r') as infile:
    reader = csv.reader(infile)
    count = 0
    for row in reader:
        if count > 0:
            if len(row) == 0:
                continue
            time = datetime.datetime.strptime(row[1],'%Y-%m-%d %H:%M:%S')
            if time.month == 3 and time.year == 2018:
                res.append(row)
        count += 1
        if count % 10000 == 0:
            print(len(res))
        
```
